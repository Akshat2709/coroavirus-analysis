```python
df=pd.read_csv(r"C:\Users\Akshat Choudhary\Desktop\covid 19\covid_19_data.csv",parse_dates=['Last Update'])
```

```python
df.rename(columns={'ObservationDate':'Date','Country/Region':'Country'},inplace=True)
```

```python
df
```

```python
df.query('Country=="India"')
```

```python
df.groupby('Date').sum()
```

```python
confirmed=df.groupby('Date').sum()['Confirmed'].reset_index()# you can check death and other things also by replacing with['confirmed']
death=df.groupby('Date').sum()['Deaths'].reset_index()  #reset index is used to put the data in the table
rec=df.groupby('Date').sum()['Recovered'].reset_index()
```

```python
confirmed
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode='lines+markers',name='Confirmed',line=dict(color='blue',width=2)))
fig.add_trace(go.Scatter(x=death['Date'],y=death['Deaths'],mode='lines+markers',name='Deaths',line=dict(color='red',width=2)))
fig.add_trace(go.Scatter(x=rec['Date'],y=rec['Recovered'],mode='lines+markers',name='Recovered',line=dict(color='green',width=2)))
```
